```python
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

```python
# Calculate the number of unique players
#Player Count
num_players = purchase_data_df["SN"].nunique()
num_players
```

```python
# present these in a data frame
total_players_df = pd.DataFrame([num_players], columns=["Total Players"])
total_players_df
```

```python
#puchasing Analysis
# Create dictionary with a key "row",list and a DataFrame

row  = {
    "Number of Unique Items" : len(purchase_data_df["Item ID"].unique()), 
    "Average Price" : "${:,.2f}".format(purchase_data_df["Price"].mean()),
    "Number of Purchases" : purchase_data_df["SN"].count(),
    "Total Revenue" : "${:,.2f}".format(purchase_data_df["Price"].sum())  
}

range_of_factors_df = pd.DataFrame(row, index=[0])
range_of_factors_df
```

```python
#confirm the number of players by gender 

gender = purchase_data_df.groupby("Gender")
gender_count = gender.nunique()["SN"]
gender_count
```

```python
# define percentage of players
#Gender Demographics 

percentage_of_players = gender_count/num_players

gender_demographics = pd.DataFrame(
    {"Total Count":gender_count,
     "Percentage of Players":percentage_of_players
    }
)

gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map("{:,.2%}".format)
gender_demographics
```

```python
#Purchasing Analysis by gender # run basic calulations to obtain purchase count, avg. purchagse prices, ave, purv=chase tital pr person etc. by gender
# create a summary data frame to hold the results 
#Optional give the displayed data clener formatting
#Display the sumary frame

purchase_count = gender["Price"].count()
average_purchase_price = gender["Price"].mean()
total_purchase_value = gender["Price"].sum()
average_total_purchase_per_person = total_purchase_value/gender_demographics["Total Count"]

purchasing_analysis = pd.DataFrame({
    "Purchase Count":purchase_count,
    "Average Purchase Price":average_purchase_price,
    "Total Purchase Value":total_purchase_value,
    "Average_Total_Purchase_per_Person": average_total_purchase_per_person})

purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Average_Total_Purchase_per_Person"] = purchasing_analysis["Average_Total_Purchase_per_Person"].map("${:,.2f}".format)
purchasing_analysis
```

```python
update_purchase_data = purchase_data_df[["SN","Age","Price","Item Name","Item ID"]]
update_purchase_data = update_purchase_data.drop_duplicates()
update_purchase_data = update_purchase_data.reset_index(drop=True)
update_purchase_data
```

```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 49]
age_ranges = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]
# age_purchasing = pdDataFrame({"Total Count":})
update_purchase_data["Age Ranges"] = pd.cut(update_purchase_data["Age"], bins, labels=bin_names, include_lowest=True)
update_purchase_data
```

```python
#age Demographics 
data_by_age_df = update_purchase_data.groupby("Age Ranges").count()[["SN"]]
data_by_age_df = data_by_age_df.rename(columns={"SN":"Total Count"})
data_by_age_df["Percentage of Players"] = data_by_age_df["Total Count"]/sum(data_by_age_df["Total Count"])
data_by_age_df["Percentage of Players"] = data_by_age_df["Percentage of Players"].map("{:,.2%}".format)
data_by_age_df
```
